## funtion to calculate


```
# This funtion used to calculate confusion matrix and accurancy and recall to both of binary classifier or mutliclass 
```



````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

def AR (x1,y1):
  X_train, X_test, y_train, y_test = train_test_split(x1,y1, test_size=0.3)
  print('Matrix shape X: ', X_train.shape)
  print('{0} samples for training'.format(X_train.shape[0]))
  print('{0} samples for testing'.format(X_test.shape[0]))
  
  log_reg = LogisticRegression()
  log_reg=log_reg.fit(X_train, y_train)

  y_pred = log_reg.predict(X_test)
  co=confusion_matrix(y_test, y_pred)
  print(co)
  acc=accuracy_score(y_test,y_pred)
  print(acc)
  accurcy=(co[0][0]+co[1][1])/(co[0][0]+co[1][1]+co[0][1]+co[1][0])
  print(accurcy)
  recall=(co[0][0]/(co[0][0]+co[0][1]))
  print(recall)
````
