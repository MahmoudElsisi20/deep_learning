# **training datasets**

````python
import os
import itertools  
import cv2
train=os.walk(training)
train_images=[]
train_dirs=[]
for (root,dirs,files) in train:
        for file in files:
          x=os.path.join(root,file)
          train_images.append(x)
        for dir in dirs:
          s=os.path.join(root,dir)
          train_dirs.append(s)
      
print(train_images)
print(train_dirs)

````

## **`actual label`**

````python
import os
import itertools  
import cv2
train=os.walk(training)
listimages=[]
for (root,dirs,files) in train:
        for file in files:
          
          x=os.path.join(root,dir)
          listimages.append(x)    
print(listimages)
actual_label1=[]
com = itertools.combinations(listimages, 2) 
v=0
for i in com:
  # print(i)
  j1=i[0]
  k1=i[1]
  if j1==k1:
    actual_label1.append(1)
  else:
    actual_label1.append(0)
  v=v+1
print(v)
print(actual_label1)
# actuallabel=np.array(actual_label0)
````

## **preprocessing**

````python
from skimage import io
from skimage import color 
from skimage.transform import resize 
import numpy as np
x=0
listofimagesafter_preprocessing=[]
for i in train_images:
  img = io.imread(i)
  image_resized = resize(img, (32, 32))
  final_image=color.rgb2gray(image_resized)
  io.imshow(final_image)
  # print('Resized Dimensions : ',final_image.shape)
  listofimagesafter_preprocessing.append(final_image)
  x=x+1
print("numberofimage",x)
# print(b)

````

## ***sum of square difference***

````python
import time
start_time = time.time()
com_train = itertools.combinations(listofimagesafter_preprocessing, 2)  
z=0
ssd=0
scores=[]
images=[]
for c in com_train:
  # print(type(c))
  # images.append(c)
  j=c[0]
  k=c[1]
  # if j.dir ==k.dir:
  #   print(1)

  scores.append(np.sum(np.square(j-k)))
  z=z+1
print("numbers of pairs",z)
print(scores)
# print(images)
# list_images=np.array(images)
# print(list_images)
````

## **predicted**

````python
sum=0
for score in scores:
  sum+=score
mean=sum/len(scores)
mean=mean-3
print(mean)
l=0
predict_label1=[]
for n in scores:
  l+=1
  if n<mean:
    predict_label1.append(1)
  else:
    predict_label1.append(0)

print(predict_label1)
print(l)
````

## **confusion_matrix**
1.   confusion_matrix
2.   accuracy_score
3.   precision_score
4.   recall_score
5.   f1_score








````python
speed_train= (time.time() - start_time)
print("speed",speed_train)
from sklearn.metrics import confusion_matrix

co=confusion_matrix(predict_label1, actual_label1)
print(co)
  
accurcy=(co[0][0]+co[1][1])/(co[0][0]+co[1][1]+co[0][1]+co[1][0])
print("ِAC",accurcy)
recall=(co[0][0]/(co[0][0]+co[0][1]))
print("r",recall)
Precision=(co[0][0]/(co[0][0]+co[1][0]))
print("P",Precision)
F_measure=(2*recall*Precision)/(recall+Precision)
print("F",F_measure)
````

# *preceptron train*

````python

x=0
listofimagesafter_preprocessing=[]
for i in train_images:
  img = io.imread(i)
  image_resized = resize(img, (32, 32))
  final_image=color.rgb2gray(image_resized)
  io.imshow(final_image)
  result = final_image.flatten()
  # print('Resized Dimensions : ',final_image.shape)
  listofimagesafter_preprocessing.append(result)
  x=x+1
print("numberofimage",x)
# print(listofimagesafter_preprocessing)

com_train = itertools.combinations(listofimagesafter_preprocessing, 2)  
z=0

features=[]
for c in com_train:
  p1=c[0]
  p2=c[1]
  f=np.concatenate((p1, p2), axis=None)
  features.append(f)
  z=z+1

print(features)
print(z)
print(len(f))

````

````python
from sklearn.neural_network import MLPClassifier
import time
start_time = time.time()
nn_clf = MLPClassifier(max_iter=100,solver='sgd',verbose='true')
nn_clf.fit(features, actual_label1)

Perceptron_pred = nn_clf.predict(features)
speed= (time.time() - start_time)
print("speed",speed)

confusion_matrixTrain=confusion_matrix(actual_label1, Perceptron_pred) 
accuracyTrain1=(confusion_matrixTrain[0][0]+confusion_matrixTrain[1][1])/(confusion_matrixTrain[0][0]+confusion_matrixTrain[0][1]+confusion_matrixTrain[1][1]+confusion_matrixTrain[1][0])
print('Classifier confusion_matrix on the training set is ' ) 
print(confusion_matrixTrain)
print('Classifier accuracy on the training set is  ' , accuracyTrain1)
````
